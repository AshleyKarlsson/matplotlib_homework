```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv("data/mouse_drug_data.csv")
clinical_trial_data = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
combined_data = pd.merge(mouse_drug_data, clinical_trial_data, on="Mouse ID")

# Display the data table for preview
combined_data.head()
```

```python
tumor_only = combined_data.drop('Metastatic Sites', axis=1)
metastatic_only = combined_data.drop('Tumor Volume (mm3)', axis=1)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_means = tumor_only.groupby(['Drug', 'Timepoint']).mean().unstack(level=0)
tumor_means.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_no_id = tumor_only.drop('Mouse ID', axis=1)
tumor_sem = tumor_no_id.groupby(['Drug', 'Timepoint']).sem().unstack(level=0)
tumor_sem.head()
```

```python
Capomulin = tumor_means.iloc[:,0]
Capomulin_sem = tumor_sem.iloc[:,0]
Infubinol = tumor_means.iloc[:,2]
Infubinol_sem = tumor_sem.iloc[:,2]
Ketapril = tumor_means.iloc[:,3]
Ketapril_sem = tumor_sem.iloc[:,3]
Placebo = tumor_means.iloc[:,5]
Placebo_sem = tumor_sem.iloc[:,5]
Timepoint = [0,5,10,15,20,25,30,35,40,45]

fig, ax = plt.subplots()

ax.errorbar(Timepoint, Capomulin, yerr=Capomulin_sem, fmt="o", color="r", 
            alpha=0.8, label="Capomulin")
ax.errorbar(Timepoint, Infubinol, yerr=Infubinol_sem, fmt="^", color="b",
            alpha=0.8, label="Infubinol")
ax.errorbar(Timepoint, Ketapril, yerr=Ketapril_sem, fmt="s", color="g",
            alpha=0.8, label="Ketapril")
ax.errorbar(Timepoint, Placebo, yerr=Placebo_sem, fmt="d", color="k",
            alpha=0.8, label="Placebo")

ax.set_xlim(-2,46)
ax.set_ylim(33,74)

ax.set_title("Tumor Response to Treatment")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_xlabel("Treatment Duration (Days)")
plt.grid()

plt.legend(loc="best", fontsize="small", fancybox=True)

plt.show()
```

```python
fig.savefig("Images/TumorResponsetoTreatment.png")
```

## Metastatic Response to Treatment

```python
# Store the Metastatic Data Grouped by Drug and Timepoint 
metastatic_means = metastatic_only.groupby(['Drug', 'Timepoint']).mean().unstack(level=0)
metastatic_means.head()
```

```python
# Store the Standard Error of Metastatic Response Grouped by Drug and Timepoint
metastatic_no_id = metastatic_only.drop('Mouse ID', axis=1)
metastatic_sem = metastatic_no_id.groupby(['Drug', 'Timepoint']).sem().unstack(level=0)
metastatic_sem.head()
```

```python
Capomulin = metastatic_means.iloc[:,0]
Capomulin_sem = metastatic_sem.iloc[:,0]
Infubinol = metastatic_means.iloc[:,2]
Infubinol_sem = metastatic_sem.iloc[:,2]
Ketapril = metastatic_means.iloc[:,3]
Ketapril_sem = metastatic_sem.iloc[:,3]
Placebo = metastatic_means.iloc[:,5]
Placebo_sem = metastatic_sem.iloc[:,5]
Timepoint = [0,5,10,15,20,25,30,35,40,45]

fig, ax = plt.subplots()

ax.errorbar(Timepoint, Capomulin, yerr=Capomulin_sem, fmt="o", color="r", 
            alpha=0.8, label="Capomulin")
ax.errorbar(Timepoint, Infubinol, yerr=Infubinol_sem, fmt="^", color="b",
            alpha=0.8, label="Infubinol")
ax.errorbar(Timepoint, Ketapril, yerr=Ketapril_sem, fmt="s", color="g",
            alpha=0.8, label="Ketapril")
ax.errorbar(Timepoint, Placebo, yerr=Placebo_sem, fmt="d", color="k",
            alpha=0.8, label="Placebo")

ax.set_ylim(-0.2, 3.7)
ax.set_xlim(-2,48)

ax.set_title("Metastatic Spread During Treatment")
ax.set_ylabel("Metastatic Sites")
ax.set_xlabel("Treatment Duration (Days)")
ax.grid()

plt.legend(loc="best", fontsize="small", fancybox=True)

plt.show()
```

```python
fig.savefig("Images/MetastaticResponsetoTreatment.png")
```

## Survival Rates

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
id_only = tumor_only.drop('Tumor Volume (mm3)', axis=1)
survival_count = id_only.groupby(['Drug', 'Timepoint']).count().unstack(level=0)
survival_count.head()
```

```python
Capomulin = survival_count.iloc[:,0]
Capomulin_percent = (Capomulin/25)*100
Infubinol = survival_count.iloc[:,2]
Infubinol_percent = (Infubinol/25)*100
Ketapril = survival_count.iloc[:,3]
Ketapril_percent = (Ketapril/25)*100
Placebo = survival_count.iloc[:,5]
Placebo_percent = (Placebo /25)*100
Timepoint = [0,5,10,15,20,25,30,35,40,45]

fig, ax = plt.subplots()

ax.plot(Timepoint, Capomulin_percent, marker="o", color="r", 
            alpha=0.8, label="Capomulin")
ax.plot(Timepoint, Infubinol_percent, marker="^", color="b",
            alpha=0.8, label="Infubinol")
ax.plot(Timepoint, Ketapril_percent, marker="s", color="g",
            alpha=0.8, label="Ketapril")
ax.plot(Timepoint, Placebo_percent, marker="d", color="k",
            alpha=0.8, label="Placebo")

ax.set_ylim(33,103)
ax.set_xlim(-2,48)

ax.set_title("Survival During Treatment")
ax.set_ylabel("Suvival Rate (%)")
ax.set_xlabel("Treatment Duration (Days)")
ax.grid()

plt.legend(loc="best", fontsize="small", fancybox=True)

plt.show()
```

```python
fig.savefig("Images/SurvivalRate.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Display the data to confirm
```

```python
tumor_sums = tumor_only.groupby(['Drug', 'Timepoint']).sum().unstack(level=0)
tumor_sums.head()
```

```python
# Percent changes in Tumor Volume
Capomulin_change = ((tumor_means.iloc[-1,0]) - (tumor_means.iloc[0,0]))/(tumor_means.iloc[0,0])*100
Infubinol_change = ((tumor_means.iloc[-1,2]) - (tumor_means.iloc[0,2]))/(tumor_means.iloc[0,2])*100
Ketapril_change = ((tumor_means.iloc[-1,3]) - (tumor_means.iloc[0,3]))/(tumor_means.iloc[0,3])*100
Placebo_change = ((tumor_means.iloc[-1,5]) - (tumor_means.iloc[0,5]))/(tumor_means.iloc[0,5])*100 
change_data = [Capomulin_change, Infubinol_change, Ketapril_change, Placebo_change]
```

```python
x = range(4)
fig, ax = plt.subplots()
bar_graph = ax.bar(x, change_data, width=1, color= ['g', 'r', 'r', 'r'])

label = np.arange(len(change_data))
ax.set_xticks(label + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))

# Function to label percentile changes
def label(drugs):
    for drug in drugs:
        value = drug.get_height()
        if value > 0:
            position = 3
        else:
            position = -5
        plt.text(drug.get_x() + drug.get_width()/2., position, '%d' % int(value)+'%',color='white', ha = 'center') 
                 
# Call function for labels on bargraph
label(bar_graph)

ax.set_title("Tumor Change over 45 Day Treatment")
ax.set_ylabel("% Tumor Volume Change")
plt.axhline(y=0, color = 'black') 
ax.grid()
```

```python
fig.savefig('Images/TumorChange.png')
```
