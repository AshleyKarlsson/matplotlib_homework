```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv("data/mouse_drug_data.csv")
clinical_trial_data = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
combined_data = pd.merge(mouse_drug_data, clinical_trial_data, on="Mouse ID")

# Display the data table for preview
combined_data.head()
```

```python
tumor_only = combined_data.drop('Metastatic Sites', axis=1)
metastatic_only = combined_data.drop('Tumor Volume (mm3)', axis=1)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_means = tumor_only.groupby(['Drug', 'Timepoint']).mean().unstack(level=0)
tumor_means.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_no_id = tumor_only.drop('Mouse ID', axis=1)
tumor_sem = tumor_no_id.groupby(['Drug', 'Timepoint']).sem().unstack(level=0)
tumor_sem.head()
```

```python
Capomulin = tumor_means.iloc[:,0]
Capomulin_sem = tumor_sem.iloc[:,0]
Infubinol = tumor_means.iloc[:,2]
Infubinol_sem = tumor_sem.iloc[:,2]
Ketapril = tumor_means.iloc[:,3]
Ketapril_sem = tumor_sem.iloc[:,3]
Placebo = tumor_means.iloc[:,5]
Placebo_sem = tumor_sem.iloc[:,5]
Timepoint = [0,5,10,15,20,25,30,35,40,45]

fig, ax = plt.subplots()

ax.errorbar(Timepoint, Capomulin, yerr=Capomulin_sem, fmt="o", color="r", 
            alpha=0.8, label="Capomulin")
ax.errorbar(Timepoint, Infubinol, yerr=Infubinol_sem, fmt="^", color="b",
            alpha=0.8, label="Infubinol")
ax.errorbar(Timepoint, Ketapril, yerr=Ketapril_sem, fmt="s", color="g",
            alpha=0.8, label="Ketapril")
ax.errorbar(Timepoint, Placebo, yerr=Placebo_sem, fmt="d", color="k",
            alpha=0.8, label="Placebo")

ax.set_xlim(-2,46)
ax.set_ylim(33,74)

ax.set_title("Tumor Response to Treatment")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_xlabel("Treatment Duration (Days)")
ax.grid()

plt.legend(loc="best", fontsize="small", fancybox=True)

plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Metastatic Data Grouped by Drug and Timepoint 
metastatic_means = metastatic_only.groupby(['Drug', 'Timepoint']).mean().unstack(level=0)
metastatic_means.head()
```

```python
# Store the Standard Error of Metastatic Response Grouped by Drug and Timepoint
metastatic_no_id = metastatic_only.drop('Mouse ID', axis=1)
metastatic_sem = metastatic_no_id.groupby(['Drug', 'Timepoint']).sem().unstack(level=0)
metastatic_sem.head()
```

```python
Capomulin = metastatic_means.iloc[:,0]
Capomulin_sem = metastatic_sem.iloc[:,0]
Infubinol = metastatic_means.iloc[:,2]
Infubinol_sem = metastatic_sem.iloc[:,2]
Ketapril = metastatic_means.iloc[:,3]
Ketapril_sem = metastatic_sem.iloc[:,3]
Placebo = metastatic_means.iloc[:,5]
Placebo_sem = metastatic_sem.iloc[:,5]
Timepoint = [0,5,10,15,20,25,30,35,40,45]

fig, ax = plt.subplots()

ax.errorbar(Timepoint, Capomulin, yerr=Capomulin_sem, fmt="o", color="r", 
            alpha=0.8, label="Capomulin")
ax.errorbar(Timepoint, Infubinol, yerr=Infubinol_sem, fmt="^", color="b",
            alpha=0.8, label="Infubinol")
ax.errorbar(Timepoint, Ketapril, yerr=Ketapril_sem, fmt="s", color="g",
            alpha=0.8, label="Ketapril")
ax.errorbar(Timepoint, Placebo, yerr=Placebo_sem, fmt="d", color="k",
            alpha=0.8, label="Placebo")

ax.set_ylim(-0.2, 3.7)
ax.set_xlim(-2,48)

ax.set_title("Metastatic Spread During Treatment")
ax.set_ylabel("Metastatic Sites")
ax.set_xlabel("Treatment Duration (Days)")
ax.grid()

plt.legend(loc="best", fontsize="small", fancybox=True)

plt.show()
```
